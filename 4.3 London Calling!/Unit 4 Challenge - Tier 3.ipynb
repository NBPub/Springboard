```python
# using yeardata as source data, index names are burough, and columns are yearly average

def create_price_ratio(burough):
    if burough not in yeardata.index:
        return 'Invalid burough name'
    
    ratio = yeardata.loc[burough, 2018] / yeardata.loc[burough, 1998]
    return ratio
```

### Back to Problem Statement . . .

**Which borough of London has seen the greatest average increase in housing prices over
the (approximately) two decades covered by the dataset?**

* will restrict analysis to the 32 buroughs
* will only use complete years of data, so from `1995` to `2022`
  * modify function written above
  * *almost 3 decades now!*
* will collect start price, end price, and ratio in a new DataFrame
  * provide plots with answer

```python
import seaborn as sns
# use for some plotting
```

```python
# again, using yeardata as source
# changing years for calculation and returning all values

def create_price_ratio2(burough):
    if burough not in yeardata.index:
        return 'Invalid burough name'
    
    start = yeardata.loc[burough, 1995]
    end = yeardata.loc[burough, 2022]
    ratio =  end/start
    return start, end, ratio
```

```python
answer = pd.DataFrame()

for burough in yeardata.index[1:33]:
    start, end, ratio = create_price_ratio2(burough)
    answer.loc[burough, 1995] = start
    answer.loc[burough, 2022] = end
    answer.loc[burough, 'ratio'] = ratio
```

```python
# decided to add raw growth later
answer['Raw'] = answer[2022] - answer[1995]
```

```python
answer.sort_values('ratio', ascending=False)
```

```python
sns.barplot(data = answer, y = answer.index, x = answer.ratio, color='k')
```

```python
answer.plot.barh(y='ratio')
```

### Modify Order of Buroughs to Reflect Ratio
* can clearly see **Hackney** has had the largest increase in this time scale

```python
df = answer.sort_values('ratio')
plt.figure(figsize=(4,8))
plt.barh(y = df.index, width=df.ratio)
plt.grid(axis='x')
plt.title('Housing Price Ratio, 2022 vs 1995')
plt.show()
```
